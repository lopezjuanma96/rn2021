```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def euler(f,x,t,dt,p):
    """
    Calcula un paso de integración del método de Euler.
    
    Argumentos de entrada:
    
        f : R^n -> R^n
        x = x(t) : R^n
        t = tiempo : R
        dt = paso de tiempo : R
        p = parametros : R^q        
        
    Retorna aproximacion numérica de
    
        x(t+dt) : R^n

    # Ejemplos:
    """
    return x + dt*f(x,t,p)
```

```python
def rk4(f,x,t,dt,p):
    """
    Calcula un paso de integración del método de Runge Kutta orden 4.
    
    Argumentos de entrada:
    
        f : R^n -> R^n
        x = x(t) : R^n
        t = tiempo : R
        dt = paso de tiempo : R
        p = parametros : R^q        
        
    Retorna aproximacion numérica de
    
        x(t+dt) : R^n

    # Ejemplos:
    """    
    k1 = f(x,t,p)
    k2 = f(x+0.5*dt*k1,t+0.5*dt,p)
    k3 = f(x+0.5*dt*k2,t+0.5*dt,p)
    k4 = f(x+dt*k3,t+dt,p)
    return x+dt*(k1+2.0*k2+2.0*k3+k4)/6.0
```

```python
def integrador_ode(m,f,x0,a,b,k,p):
    """
    Integra numéricamente la ODE
    
        dx/dt = f(x,t)
        
    sobre el intervalo t:[a,b] usando k pasos de integración y el método m, bajo condicion inicial x(a)=x0.
    No es necesario que a<b.
    
    Argumentos de entrada:
    
        m = metodo de integracion (ej. euler, rk2, etc.)
        f : R^n -> R^n
        x0 = condicion inicial : R
        a = tiempo inicial : R
        b = tiempo final : R
        k = num. pasos de integracion : N
        p = parametros : R^q        
    
    Retorna:
    
        t : R^{k+1} , t_j = a+j*dt para j=0,1,...,k
        x : R^{k+1,n} , x_ij = x_i(t_j) para i=0,1,...,n-1 y j=0,1,...,k
        
    donde a+k*dt = b.
    """  
    assert k>0
    n = len(x0)
    dt = (b-a)/k
    x = np.zeros((n,k+1))
    t = np.zeros(k+1)
    x[:,0] = x0
    t[0] = a
    for j in range(k):
        t[j+1] = t[j] + dt
        x[:,j+1] = m(f,x[:,j],t[j],dt,p)
    return t,x
```

```python
def lin_func(x, t, p):
    return -p[0]*x+p[1]

t_array, x_array = integrador_ode(rk4, lin_func, [1], 0, 5, 20, (2, 3))

print(t_array)
print(x_array)
```

```python
def sol_func(x0, a, b, k, params):
    t = np.linspace(a, b, k)
    return t, (x0 - params[1]/params[0])*np.exp(-params[0]*t)+params[1]/params[0]

t_array, x_array = sol_func(1, 0, 5, 20, (2, 3))
print(t_array)
print(x_array)
```

5. Para representar la funcion de $\dot{x}$ en funcion de x usaremos 100 valores de x. 

```python
def plot_x_dotx(x_range, fixed_points_tolerance, function, function_args):
    if len(x_range) > 2:
        x = np.arange(x_range[0], x_range[1], x_range[2])
    else:
        x = np.arange(x_range[0], x_range[1])
    
    dotx = function(x, *function_args)
    
    fixed_points = [[], [], []]
    arrows_r = []
    arrows_l = []
    for d in range(dotx.shape[0]):
        if abs(dotx[d]) <= fixed_points_tolerance:
            if dotx[d-1] * dotx[d+1] < 0:
                if dotx[d-1] > 0:
                    fixed_points[0].append(x[d])
                else:
                    fixed_points[1].append(x[d])
            else:
                fixed_points[2].append(x[d])
        else:
            if dotx[d] < 0:
                arrows_l.append(x[d])
            else:
                arrows_r.append(x[d])
                
    print("Fixed Points:", fixed_points)
    
    plt.figure()
    plt.title("X vs dX/dt")
    
    #axis
    plt.axhline(0, color="black", alpha=0.3)
    plt.axvline(0, color="black", alpha=0.3)
    
    #main plot
    plt.plot(x, dotx, color="black")
    
    #dotx = 0 line
    plt.axhline(0, color="grey", linestyle = "--")
    
    #fixed_points
    plt.scatter(fixed_points[0], [0]*len(fixed_points[0]), color="black")
    plt.scatter(fixed_points[1], [0]*len(fixed_points[1]), color="#DCDCDC") #grey instead of white
    plt.scatter(fixed_points[2], [0]*len(fixed_points[2]), color="red")
    
    #arrows
    plt.plot(arrows_r, [0]*len(arrows_r), color="blue", marker=">", linestyle = "")
    plt.plot(arrows_l, [0]*len(arrows_l), color="blue", marker="<", linestyle = "")
    
    return x, dotx

_, _ = plot_x_dotx((-1, 3, 0.1), 0.001, lin_func, (None, (2,3)))   
```

- de $N=0$ a $N=K$ positiva.

3. Runge Kutta 4to Orden

```python
def cuad_func(n, t, p):
    return p[0]*n*(1-n/p[1])

r = 1
k = 1

n0_values = [0.1*k, 0.7*k, 1.2*k]
t_array, n_array = integrador_ode(rk4, cuad_func, n0_values, 0, 5, 100, (r, k))

print(t_array)

for n0 in range(len(n0_values)):
    print("\nFor N0 =", n0_values[n0], "\n")
    print(n_array[n0, :])
```

```python
for n0 in range(len(n0_values)):
    plt.plot(t_array, n_array[n0, :], label="N0="+str(n0_values[n0]))
    
plt.legend()
```

4. Para graficar con el integrador usaremos valores de N0 entre -5 y 5

```python
n0_values = list(np.arange(0, 1, 0.1))
t_array, n_array = integrador_ode(rk4, cuad_func, n0_values, 0, 5, 100, (r, k))
```

```python
for n0 in range(len(n0_values)):
    plt.plot(t_array, n_array[n0, :], label="N0="+str(round(n0_values[n0], 2)))
    
plt.legend()
```

No usamos valores negativos porque generan overflow

Solucion exacta: La solucion exacta obtenida surge de suponer K = 1 para todos los casos, ver "Intermezzo" en solución.

```python
def sol_func(x0, a, b, k, params):
    t = np.linspace(a, b, k)
    return t, ((x0/(1-x0))*np.exp(params[0]*t))/(1+((x0/(1-x0))*np.exp(params[0]*t)))

for x0 in n0_values:
    t, x = sol_func(x0, 0, 5, 100, [r])
    plt.plot(t, x, label="N0="+str(round(x0,2)), linestyle = "--")
    
plt.legend()
```
